```python
# Importação bibliotecas
!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
#Passo 1: Copiar informações do site.

# Abrir o navegador 
navegador = webdriver.Chrome()
# Entrar no Google ou Site
navegador.get("https://google.com.br/")

# Pesquisar Dolar §Comands:.send_keys():'escrever no elemento' .click():'clicar no elemento' .get_attribute(): 'pegar informações do elemento'§

navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input' ).send_keys("cotacao dolar")
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input' ).send_keys(Keys.ENTER)
# Copiar informação
Cotacao_Dolar = navegador.find_element('xpath','//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

# Pesquisar Euro

navegador.get("https://google.com.br/")
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input' ).send_keys("cotacao euro")
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input' ).send_keys(Keys.ENTER)
Cotacao_Euro = navegador.find_element('xpath','//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

# Pesquisar Ouro

navegador.get("https://www.melhorcambio.com/ouro-hoje")
Cotacao_Ouro = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute("value")
Cotacao_Ouro = Cotacao_Ouro.replace(",", ".")
print(Cotacao_Dolar)
print(Cotacao_Euro)
print(Cotacao_Ouro)

navegador.quit()
```

```python
import pandas as pd
#Passo 2: Importar banco de dados no excel
tabela = pd.read_excel("Banco_Dados_Teste.xlsx")
display(tabela)
```

```python
#Passo 3: Atualizar Banco de Dados

# Substituir dados tabela.l]=="Dolar", "Cotacao") = float(informacao01)oc(linha, coluna) = float§numero decimal§(informacao01)
tabela.loc[tabela["Moeda"] == "Dólar", "Cotação"] = float(Cotacao_Dolar)
tabela.loc[tabela["Moeda"] == "Euro", "Cotação"] = float(Cotacao_Euro)
tabela.loc[tabela["Moeda"] == "Ouro", "Cotação"] = float(Cotacao_Ouro)

# atualizar o preço base reais (preço base original * cotação)
tabela["Preço de Compra"] = tabela["Preço Original"] * tabela["Cotação"]

# atualizar o preço final (preço base reais * Margem)
tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela["Margem"]

# tabela["Preço de Venda"] = tabela["Preço de Venda"].map("R${:.2f}".format)

display(tabela)
```
